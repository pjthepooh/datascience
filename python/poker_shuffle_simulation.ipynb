# Simulation

```python
import matplotlib.pyplot as plt
% matplotlib inline
```

```python
from poker_shuffle_simulation import PokerShuffleSimulator, NUM_SHUFFLES
```

```python
sim_bottom_up = PokerShuffleSimulator(method='bottom_up')
sim_bottom_up.simulate_shuffle()
sim_bottom_up.evaluate_simulation()
error_avg_bottom_up = sim_bottom_up.eval_summary[0]
```

```python
sim_complete_riffle = PokerShuffleSimulator(method='complete_riffle')
sim_complete_riffle.simulate_shuffle()
sim_complete_riffle.evaluate_simulation()
error_avg_complete_riffle = sim_complete_riffle.eval_summary[0]
```

```python
sim_random_riffle = PokerShuffleSimulator(method='random_riffle')
sim_random_riffle.simulate_shuffle()
sim_random_riffle.evaluate_simulation()
error_avg_random_riffle = sim_random_riffle.eval_summary[0]
```

```python
n_shuffles = range(1, NUM_SHUFFLES + 1)
frontsize=15

plt.style.use('seaborn-darkgrid')

fig, ax = plt.subplots(figsize=(12, 6))
ax.plot(n_shuffles, error_avg_bottom_up, 'o-', linewidth=3, markersize=5, label='Bottom-up',color='#EE3B3B')
ax.plot(n_shuffles, error_avg_complete_riffle, 'o-', linewidth=3, markersize=5, label='Complete Riffle', color='#1E90FF')
ax.plot(n_shuffles, error_avg_random_riffle, 'o-', linewidth=3, markersize=5, label='Random Riffle', color='#76EE00')
ax.legend(loc='best', fontsize=frontsize)
plt.title('Poker Shuffle Simulation', fontsize=frontsize)
plt.xlabel('Number of Shffles', fontsize=frontsize)
plt.ylabel('Error', fontsize=frontsize)
ax.grid(True)
```
